## Corrected — file is inside `data/raw/BIG5/`

```python
df = pd.read_csv("../data/raw/BIG5/data.csv", sep="\t")
total_loaded = len(df)
```

```python
df_filt = df[(df["age"] >= 13) & (df["age"] <= 80)]
n_kept = len(df_filt)
```

```python
fig, ax = plt.subplots()
ax.hist(df_filt["age"], bins=30)
ax.set_xlabel("Age (years)")
ax.set_ylabel("Count")
ax.set_title(f"BIG5 age distribution (n={n_kept})")

out_path = "../reports/age_distribution.png"
plt.savefig(out_path, dpi=150)
plt.show()
```
